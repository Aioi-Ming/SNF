```python
def plot_potential(model,labels=True, cbar=True, figsize=(5, 5)):
    # 2D potential
    xgrid = np.linspace(-1.5, 1.5, 100)
    ygrid = np.linspace(-4, 4, 100)
    Xgrid, Ygrid = np.meshgrid(xgrid, ygrid)
    X = np.vstack([Xgrid.flatten(), Ygrid.flatten()]).T
    E = model.energy(X)
    E = E.reshape((100, 100))
    E = np.minimum(E, 10.0)
    plt.figure(figsize=figsize)
    plt.contourf(Xgrid, Ygrid, E, 50, cmap='jet', vmax=4)
    if cbar:
        cbar = plt.colorbar()
        cbar.set_label('Energy / kT', labelpad=-15, y=0.6)
        cbar.set_ticks([-10, -5, 0, 5, 10])
    if labels:
        plt.xlabel('$x_1$ / a.u.')
        plt.ylabel('$x_2$ / a.u.')
    else:
        plt.xticks([])
        plt.yticks([])   
```

### Explore

```python
model=Cosine(w=2)
```

```python
_=model.plot_dimer_energy()
```

```python
nsteps = 500000
stride = 10
noise=0.05
burnin=100000
xl=np.array([-0.5,0])
xr=np.array([0.5,0])
```

```python
sampler_l = MetropolisGauss(model, xl, noise=noise, burnin=burnin, stride=stride)
sampler_l.run(nsteps)
```

```python
sampler_r = MetropolisGauss(model, xr, noise=noise, burnin=burnin, stride=stride)
sampler_r.run(nsteps)
```

```python
xl=sampler_l.traj
xr=sampler_r.traj
```

```python
x_brute=np.vstack((xl,xr))
x_brute.shape
```

```python
_=model.plot_sample_energy(x_brute)
```

```python
x0 = x_brute[100:101]
X0 = np.repeat(x0, 1000, axis=0)
X0 += 0.01 * np.random.randn(X0.shape[0], X0.shape[1])
```

```python
plot_potential(model,labels=False, cbar=False, figsize=(5, 5))
plt.plot(x0[:, 0], x0[:, 1], color='white', linewidth=0, marker='+', markersize=20, markeredgewidth=3)
```

```python
model_SNF=BG_RealNVP(model,dim=model.dim,stochastic=False,step_size=0.05, nsteps=5, n_hidden=256, n_block=8)
```

```python
Explore_model=Configuration(model_SNF,X0.astype('float32'),capacity=10000)
```

```python
ax=plt.gca()
ax.hist2d(Explore_model.X[:,0],Explore_model.X[:,1],bins=100,norm=matplotlib.colors.LogNorm())
plt.show()
```
